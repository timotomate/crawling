# 크롤링

## 크롤링

```python
from bs4 import BeautifulSoup
from selenium import webdriver
import time
driver = webdriver.Chrome("./chromedriver")
```

```python
result_list = []
for i in range(1, 8):
    driver.get("https://www1.president.go.kr/petitions/best?page=" + str(i))
    soup = BeautifulSoup(driver.page_source, 'html.parser')
    for i in soup.select("#cont_view > div.cs_area > div > div > div.board.text > div.b_list.category > div.bl_body > ul > li"):
        print(i.find("div", class_="bl_subject").text[3:].strip())
        result_list.append(i.find("div", class_="bl_subject").text[3:].strip())
    time.sleep(2)
```

```python
driver.close()
```

## 크롤링 한 데이터를 엑셀로 저장

```python
from openpyxl import Workbook

write_workbook = Workbook()
write_cell = write_workbook.active

for i in range(1, len(result_list) + 1):
    write_cell.cell(i, 1, result_list[i-1])
    
write_workbook.save("bluehouse.xlsx")
```

# 자료 불러오기 openpyxl

# 국민청원 제목 수집 후, 명사들만 추출

```python
from konlpy.tag import Kkma
import collections
```

```python
kkma = Kkma()
```

```python
list_temp = []
```

```python
for row in list_excel:
    print(kkma.nouns(row))
    list_temp += kkma.nouns(row)
```

```python
print(list_temp)
```

## 1글짜 짜리 명사는 제외시키기

```python
result_list = []

for check in list_temp:
    if len(check) != 1:
        result_list.append(check)
```

```python
result_list
```

```python
collections.Counter(result_list)
```

```python
collections.Counter(result_list).most_common(10)
```

## 예시, 미국 헌법으로 word cloud 만들어보기

```python
from wordcloud import WordCloud
```

```python
#us2 = https://constitutionus.com/ 여기 페이지 그냥 일부 복붙해서 us2.txt로 저장함
text = open('us2.txt', 'rt', encoding = 'UTF8').read()
wordcloud = WordCloud().generate(text)

plt.figure(figsize=(12,12))
plt.imshow(wordcloud, interpolation = "bilinear")
#plt.axis("off") 
plt.show()
```

```python
list_result = []
for check in list_temp:
    if len(check) != 1:
        print(check)
        list_result.append(check)
```

```python
# 리스트 형태를 문자열로 변경
last_text=""
for i in list_result:
    last_text = last_text +" "+i
```

```python
last_text
```

```python
"""
font_path에서 오류나는 경우 많은데 다필요없고, 그냥 이 주피터 파일과 같은 위치에
폰트파일 복붙하면 됨.
"""

wordcloud = WordCloud(font_path="NanumGothic.ttf").generate(last_text)
```

```python
wordcloud.words_
```

```python
plt.figure(figsize=(12,12))
plt.imshow(wordcloud, interpolation = "bilinear")
plt.axis("off") 
plt.show()
```

```python
from os import path
from PIL import Image
import numpy as np
import os

d = path.dirname(__file__) if "__file__" in locals() else os.getcwd()


mask = np.array(Image.open(path.join(d, 'cloud.png')))
```
